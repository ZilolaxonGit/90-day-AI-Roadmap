```python
# Problem 5

friends = {
    "first_friend" : {
        "name" : "Zilola",
        "age" : 24,
    },
    "second_friend" : {
        "name" : "Iroda",
        "age" : 28,
    },
    "third_friend" : {
        "name" : "Malika",
        "age" : 15,
    }
}

for i in friends.keys():
    print(i)

for  value in friends.values():
    print(value["name"])
    print(value["age"])
```

```python
# Task 6
friends = {
    "first_friend" : {
        "name" : "Zilola",
        "age" : 24,
    },
    "second_friend" : {
        "name" : "Iroda",
        "age" : 28,
    },
    "third_friend" : {
        "name" : "Malika",
        "age" : 15,
    }
}
for i in friends:
    for key, val in friends[i].items():
        print(f"{key}: {val}")
```

```python
# Task 7
text = "I really really love AI world"

word_count = {}

sent = text.split()

for i in sent:
    if i in word_count:
        word_count[i] += 1
    else:
        word_count[i] = 1
print(word_count)
```

```python
# Task 8
number = 25

dictionary = {
    "first_number" : 10,
    "second_number" : 30,
    "third_number" : 45,
}

for i in dictionary:
    if dictionary[i] >= number:
        print(i)
```

```python
# Task 9

dictionary = {}

for i in range(1,6):
    dictionary[i] = i**2
print(dictionary)
```
