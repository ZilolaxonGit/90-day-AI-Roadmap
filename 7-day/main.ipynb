```python
# Student Score Manager

students = {}

def view_students():
    num = 1
    for i in students:
        print(f"{num}. {i}:{students[i]} score.\n")
        num += 1

def add_student(name:str, score:float):
    students[name] = score

def update_scores(name:str, score: float):
    if name in students:
        students[name] = score
        print("Student's score is updated successfully")

    else:
        print("Student not found")

def calculate_avg_scores():
    students_len = len(students)
    students_score_sum = sum(students.values())
    return students_score_sum / students_len if students_len > 0 else 0

def top_student():
    if students:
        max_score = max(students.values())
        for i, val in students.items():
            if val == max_score:
                return f"Top student is {i} with {val} score"
    else:
        return "Not any student found"
while True:
    options = """
            Choose one of them:
            1. Add student
            2. Update student score
            3. View all students
            4. Calculate avg scores
            5. Find top student
            6. Quit a program
        """
    option = int(input(options))

    if option == 1:
        print("Add new student\n")
        name = input("Student name: ")
        score = float(input("Student Score: "))
        add_student(name, score)
    
    elif option == 2:
        print("Update score of student\n")
        name = input("Student name: ")
        score = float(input("New Student Score: "))
        update_scores(name, score)

    elif option == 3:
        print("All studnts list\n")
        view_students()
    
    elif option == 4:
        avg_score = calculate_avg_scores()
        print(f"Students avh score is {avg_score}")
    
    elif option == 5:
        print(top_student())

    elif option == 6:
        print("You quited the program")
        break
    else:
        print("The inalid operation number. Choose again.")

    
```
