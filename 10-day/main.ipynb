```python
import numpy as np

arr = np.array([2, 4, 4, 5,])

mean = np.mean(arr)
variance = np.var(arr)
std_dev = np.std(arr)

print("Mean:", mean)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```
