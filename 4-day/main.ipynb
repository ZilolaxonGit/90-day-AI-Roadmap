```python
tasks = """

Create a list of 5 of your favorite colors and print each color.

Print the first three items of the list using slicing.

Print the last two items of the list.

Reverse the list using slicing.

Add a new color to the end of the list.

Insert a new color at index 2.

Remove a color by name.

Check if "blue" is in your list.

Print the list length.

Loop through the list and print items with their index.

"""
```

```python
# 1-task

colors = ["pink", 'blue', "green", "yellow", "aqua"]

for i in colors:
    print(i)
```

```python
# 2-problem
colors = ["pink", 'blue', "green", "yellow", "aqua"]
print(colors[:3])
```

```python
# 3-problem
colors = ["pink", 'blue', "green", "yellow", "aqua"]

print(colors[:2:-1])
```

```python
# 4-problem
colors = ["pink", 'blue', "green", "yellow", "aqua"]

print(colors[::-1])
```

```python
# 5-problem
colors = ["pink", 'blue', "green", "yellow", "aqua"]

colors.append("grey")
print(colors)
```

```python
# 6-problem
colors = ["pink", 'blue', "green", "yellow", "aqua"]
colors.insert(2, "purple")
print(colors)
```

```python
# 7-problem
colors = ["pink", 'blue', "green", "yellow", "aqua"]

colors.remove("pink")
print(colors)
```

```python
# 8-problem
colors = ["pink", 'blue', "green", "yellow", "aqua"]

print("blue" in colors)
```

```python
# 9-problem
colors = ["pink", 'blue', "green", "yellow", "aqua"]
print(len(colors))
```

```python
# 10-problem
colors = ["pink", 'blue', "green", "yellow", "aqua"]

for i, val in enumerate(colors):
    print(f"{i+1}:{val}")
```
