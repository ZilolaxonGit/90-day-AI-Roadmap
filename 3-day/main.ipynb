```python
tasks = """

10 Simple Loop Practice Problems


Problem 1

Print numbers from 1 to 10.

Problem 2

Print even numbers between 1 and 20.

Problem 3

Print the first 10 squares.

Problem 4

Print each item in a list with its index (use enumerate).

Problem 5

Sum numbers from 1 to 100.

Problem 6

Ask user for 5 names and print them.

Problem 7

Print numbers 1 to 10, but skip 5 (use continue).

Problem 8

Print numbers 1 to 10, but stop at 7 (use break).

Problem 9

Loop through a list of grades and print "Pass" if grade ≥ 50, else "Fail".

Problem 10

Ask the user to guess a number between 1 and 10 until correct (use while loop).

"""
```

```python
# Problem 1

for i in range(1, 11):
    print(i)
```

```python
# Problem 2

for i in range(2, 21, 2):
    print(i)
```

```python
# Problem 3

for i in range(1, 11):
    print(i ** 2)
```

```python
# Problem 4
my_list = ["Iroda", "Zilola", "Malika", "Muhammadjon"]
for index, value in enumerate(my_list):
    print(f"{index + 1}. {value}")
```

```python
# Problem 5

print(sum(range(1, 101)))
```

```python
# Problem 6

first_name = input("first name: ")
second_name = input("second name: ")
third_name = input("third name: ")
fourth_name = input("fourth name: ")
fifth_name = input("fifth name: ")


print(first_name, second_name, third_name, fourth_name, fifth_name)
```

```python
# Problem 7

for i in range(1, 11):
    if i == 5:
        continue
    print(i)
```

```python
# Problem 8

for i in range(1, 11):
    if i == 7:
        break
    print(i)
```

```python
# 9-problem

grades = [12, 55, 67, 89, 100, 4, 45]

for i in grades:
    if i >= 55:
        print("You passed")
    else:
        print("You failed")
```

```python
# 10-problem

number = 10
user_num = int(input("Guess the number from 0 to 10"))

# First way to solve
while True:
    if user_num == number:
        print("You guessed right, congratulations :)")
        break
    else:
        print("You didn't guess correct, please try again")
        user_num = int(input("Guess the number from 0 to 10"))
```

```python
# 10-problem

number = 10
user_num = int(input("Guess the number from 0 to 10"))

# First way to solve
while user_num != number:
    print("You didn't guess correct, please try again")
    user_num = int(input("Guess the number from 0 to 10"))
print("You guessed correctly, congratulations :)")
```
