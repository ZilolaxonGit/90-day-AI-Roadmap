```python
tasks = """

Write a function greet_user(name) → prints "Hello {name}!"

Write a function add_numbers(a, b) → returns sum

Write a function square(n) → returns n squared

Write a function is_even(n) → returns True if even, False if odd

Write a function calculate_area(radius) → returns area of a circle

"""
```

```python
# Task 1

def greet_person_by_name(name:str = "Anonymous user"):
    return f"Hello, {name}, how are you ?"

print(greet_person_by_name("Zilola"))
print(greet_person_by_name())
print(greet_person_by_name(12))
```

```python
# Task 2
def add_numbers(num1: int, num2:int):
    return sum((num1, num2))

print(add_numbers(1,2))
```

```python
# Task 3

def square(n:int):
    return n**2
print(square(5))
```

```python
# Task 4

def is_even(num:int):
    return num % 2 == 0

print(is_even(2))
print(is_even(3))
```

```python
# Task 5
import math
def calculate_area(radius: float):
    return math.pi * (radius ** 2)

print(calculate_area(5))
```
